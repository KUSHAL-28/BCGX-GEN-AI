#### **Overview of Financial Performance (2021–2023)**

Each company’s performance is evaluated across several dimensions: total revenue, net income, total assets, total liabilities, and cash flow from operating activities. This data sheds light on the financial health and strategic positioning of Microsoft, Tesla, and Apple over the given period.

```python
import pandas as pd
import numpy as np
data = pd.read_csv('/content/Financial_Data.csv')
df = pd.DataFrame(data)
df
```

percentage change from one year to the next

```python
df['Revenue Growth (%)'] = df.groupby('Company')['Total Revenue'].pct_change() * 100
df['Net Income Growth (%)'] = df.groupby('Company')['Net Income'].pct_change() * 100
```

```python
df
```

```python
import matplotlib.pyplot as plt

# Create a figure with multiple subplots
fig, axes = plt.subplots(2, 2, figsize=(12, 12))

# Plot 1: Total Revenue
revenue_by_year = df.groupby(['Company', 'Year'])['Total Revenue'].sum().unstack()
revenue_by_year.plot(kind='bar', ax=axes[0, 0])
axes[0, 0].set_title('Total Revenue of Companies Over the Years')
axes[0, 0].set_xlabel('Company')
axes[0, 0].set_ylabel('Total Revenue')
axes[0, 0].legend(title='Year')

# Plot 2: Net Income
revenue_by_year = df.groupby(['Company', 'Year'])['Net Income'].sum().unstack()
revenue_by_year.plot(kind='bar', ax=axes[0, 1])
axes[0, 1].set_title('Net Income of Companies Over the Years')
axes[0, 1].set_xlabel('Company')
axes[0, 1].set_ylabel('Net Income')
axes[0, 1].legend(title='Year')

# Plot 3: Total Assets
revenue_by_year = df.groupby(['Company', 'Year'])['Total Assets'].sum().unstack()
revenue_by_year.plot(kind='bar', ax=axes[1, 0])
axes[1, 0].set_title('Total Assets of Companies Over the Years')
axes[1, 0].set_xlabel('Company')
axes[1, 0].set_ylabel('Total Assets')
axes[1, 0].legend(title='Year')

# Plot 4: Total Liabilities
revenue_by_year = df.groupby(['Company', 'Year'])['Total Liabilities'].sum().unstack()
revenue_by_year.plot(kind='bar', ax=axes[1, 1])
axes[1, 1].set_title('Total Liabilities of Companies Over the Years')
axes[1, 1].set_xlabel('Company')
axes[1, 1].set_ylabel('Total Liabilities')
axes[1, 1].legend(title='Year')

# Adjust layout and display
plt.tight_layout()
plt.show()
```

```python
df['Assets Growth (%)'] = df.groupby('Company')['Total Assets'].pct_change() * 100
df['Liabilities Growth (%)'] = df.groupby('Company')['Total Liabilities'].pct_change() * 100
df['Cash Flow from Operations Growth(%)'] = df.groupby('Company')['Cash Flow from Operating Activities'].pct_change() * 100
```

```python
df
```

#### **Yearly Growth Analysis**

The annual growth in revenue, net income, assets, liabilities, and cash flow is examined for each company. The results reveal varied performance trends across companies.

**Revenue Growth (2021–2023):**
- **Microsoft** showed a modest revenue growth, with a -6.4% decline from 2022 to 2023, following a 15.2% rise from 2021 to 2022.
- **Tesla** experienced significant growth, notably a 15.8% increase from 2021 to 2022, though growth slowed in 2023.
- **Apple** demonstrated steady revenue growth, with a peak in 2022 (+2.88%) but a slight contraction of 7.2% in 2023.

**Net Income Growth:**
- **Microsoft** maintained strong net income stability, despite a slight decrease of -15.76% in 2021.
- **Tesla** exhibited high volatility in net income, seeing a major increase (+56.04%) from 2021 to 2022, then facing a decline of 16.14% in 2023.
- **Apple** remained consistent, with minor fluctuations (e.g., +2.89% in 2022).


```python
df = df.fillna(0)
df
```

- **Net Income**: Net income grew from \$161,470 million in 2021 to a peak of \$185,097 million in 2022, with a slight decline to \$184,330 million in 2023, which could suggest increased expenses or investments affecting profitability.

#### **Correlation Between Revenue and Net Income**

A strong positive correlation (0.945) between revenue and net income indicates that, across these companies, increases in revenue tend to lead to proportional increases in net income. This correlation suggests that these comapanies maintain effective cost management and profitability strategies, benefiting their bottom lines as revenues grow.

#### **Median Asset and Liability Growth (2021–2023)**
Median growth rates in assets and liabilities offer insights into the stability and capital structure management of each company. Notably:
- **Decline in 2021 and 2022**: Both assets and liabilities showed negative median growth rates, possibly reflecting economic challenges, capital restructuring, or debt reduction.
- **Stabilization in 2023**: Growth rates in 2023 were near-zero, suggesting a period of stabilization and balanced management of assets and liabilities.

```python
# Calculate average revenue and net income for each company over the years
avg_revenue_by_company = df.groupby('Company')['Total Revenue'].mean()
avg_net_income_by_company = df.groupby('Company')['Net Income'].mean()

print("Average Revenue by Company:\n", avg_revenue_by_company)
print("\nAverage Net Income by Company:\n", avg_net_income_by_company)

# Calculate the total revenue and net income for each year across all companies
revenue_by_year = df.groupby('Year')['Total Revenue'].sum()
net_income_by_year = df.groupby('Year')['Net Income'].sum()

print("\nTotal Revenue by Year:\n", revenue_by_year)
print("\nTotal Net Income by Year:\n", net_income_by_year)

# Group by year and find the median asset and liability growth
median_growth_by_year = df.groupby('Year').agg({
    'Assets Growth (%)': 'median',
    'Liabilities Growth (%)': 'median'
})

print("\nMedian Asset and Liability Growth by Year:\n", median_growth_by_year)

# Calculate the average growth rates over the period
avg_growth_rates = df.groupby('Company')[['Revenue Growth (%)', 'Net Income Growth (%)']].mean()

print("Average Growth Rates:\n", avg_growth_rates)

# Analyze the correlation between revenue growth and net income growth
correlation_growth = df['Revenue Growth (%)'].corr(df['Net Income Growth (%)'])
print("\nCorrelation between Revenue Growth and Net Income Growth:", correlation_growth)
```
