```python
data=pd.read_csv('Final_Data.csv')
df=pd.DataFrame(data)
df
```

```python
import pandas as pd

data = {}

# Iterate over unique companies
for company in df['Company'].unique():
    data[company] = {}
    company_data = df[df['Company'] == company]  # Filter data for the current company

    # Iterate over years
    for year in company_data['Year'].unique():
        year_data = company_data[company_data['Year'] == year].iloc[0]  # Get data for the year
        data[company][year] = year_data.drop(['Company', 'Year']).to_dict() # Include all other columns

print(data)
def financial_chatbot(company, year, metric):
    # Validate company and year
    if company in data and year in data[company]:
        # Retrieve the metric
        company_year_data = data[company][year]
        metric_value = company_year_data.get(metric)

        # Check if the metric exists for the given year
        if metric_value:
            return f"\n******************The {metric} for {company} in {year} is: {metric_value}**********\n"
        else:
            return f"Sorry, we don’t have data for '{metric}' for {company} in {year}."
    else:
        return f"Sorry, we don’t have data for {company} in {year}."

# Main loop to interact with the chatbot
print("Welcome to the AI-powered financial chatbot!")
print("We have fiscal data from 2021 to 2023 for Microsoft, Tesla, and Apple.")

while True:
    # Take user input
    company = input("Which company's data do you want?: ")
    year = int(input("Of which year?: "))

    print("1. Total Revenue")
    print("2. Total Net Income")
    print("3. Total Assets")
    print("4. Total Liabilities")
    print("5. Cash Flow from Operating Activities")
    print("6. Revenue Growth (%)")
    print("7. Net Income Growth (%)")

    metric_options = {
        "1": "Total Revenue",
        "2": "Total Net Income",
        "3": "Total Assets",
        "4": "Total Liabilities",
        "5": "Cash Flow from Operating Activities",
        "6": "Revenue Growth (%)",
        "7": "Net Income Growth (%)"
    }

    metric_choice = input("Enter the metric's number you want to know: ")
    metric = metric_options.get(metric_choice)

    # Call the chatbot function
    if metric:
        print(financial_chatbot(company, year, metric))
    else:
        print("Invalid choice. Please try again.")

    # Ask if the user wants to continue
    continue_choice = input("Do you want to continue (yes/no)?: ")
    if continue_choice.lower() == "no":
        break

print("Thank you!")
```
